Importing Linear model from sklearn.

```python
from  sklearn  import  linear_model
```

```python
reg = linear_model.LinearRegression()

X = df[ [ 'Runs','BF' ] ] 

Y = df[ '4s' ]
```

Here,

X = Dataframe containing features (Runs,BF)


Y = Dataframe containing Target (Number of 4s)

Importing Train test split to split the dataset using train_test_split utility from sklearn into training and 
testing with the number of 4s as a target.

```python
from sklearn.model_selection import train_test_split
```

Spliting the given datasets in ration of 80% train and 20% test.

```python
X_train , X_test , Y_train , Y_test = train_test_split( X , Y , test_size = 0.2 )
```

Training the model using X_train and Y_train datasets.

```python
reg.fit( X_train , Y_train )
```

here,

l = Lenght of the test dataset

r = Predicted values of the number of 4s for the test data.

sub = Difference between predicted and original values of number of 4s for test data.

sq = Square of difference between predicted and original values of number of 4s for test data.

```python
l = len( X_test )

r = reg.predict( X_test )

sub = np.subtract( Y_test , r )

sq = np.power( sub , 2 )
```

We need to calculate the mean square error for the train data predictions and for the test data predictions

so,

se = Sum of squares of difference between predicted and original values of number of 4s for test data.

and

mse = Mean square error for the train data predictions and test data predictions

```python
se = 0

for i in sq:
    se+=i

mse = se/l

print( mse )
```

So, The value of mean square error is found.
